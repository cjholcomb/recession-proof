```python
def import_data(year):
    filepath = 'data/' + year + '.csv'
    return pd.read_csv(filepath, dtype = schema_dict)
```

```python
def append_years(df, years):
    for year in years:
        #if year != '2007':
        filepath = 'data/' + year + '.csv'
        df = df.append(pd.read_csv(filepath, dtype=schema_dict), ignore_index = True)
    return df
```

```python
df = append_years(df, recession_years)
```
